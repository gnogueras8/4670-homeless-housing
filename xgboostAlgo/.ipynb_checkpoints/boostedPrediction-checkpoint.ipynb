```python
import xgboost as xgb
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../final_data_dummy.csv')
```

```python
df.describe()
```

```python
y = df.pcthmlsofciv
```

```python
x = df.drop('pcthmlsofciv', axis = 1 )
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state=123)
```

```python
xg_cl = xgb.XGBClassifier(objective='logistic', seed=123)
xg_cl.fit(x_train, y_train)
```

```python
xg_cl.score(x_train, y_train)
```
